```python
import requests

url = r'https://catfact.ninja/'
method = 'facts'
payload = {'limit': 1000}
r = requests.get(url+method, params=payload)
r.json()
```

```python
def dictToFacts(json):
    facts = []
    for item in json['data']:
        facts.append(item['fact'])
    return facts

data = dictToFacts(r.json())
data_str = ''
for fact in data:
    data_str += fact + '\n'

data_file = 'catfacts.txt'

f = open(data_file, 'w')
f.write(data_str)
f.close()
```
